```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/friends_info.csv')
```

```python
print(df.head(25))
```

### **Data Modelling**

```python
characters = ['Chandler', 'Joey', 'Ross', 'Monica', 'Rachel', 'Phoebe']
```

```python
def check_character_in_title(title, character):
    return 1 if character in title else 0
```

```python
for character in characters:
    df[character] = df['title'].apply(lambda x: check_character_in_title(x, character))
```

```python
print(df.head(25))
```

```python
chandler_df = df[df['Chandler'] == 1]
print(chandler_df)
```

```python
joey_df = df[df['Joey'] == 1]
print(joey_df)
```

```python
ross_df = df[df['Ross'] == 1]
print(ross_df)
```

```python
monica_df = df[df['Monica'] == 1]
print(monica_df)
```

```python
rachel_df = df[df['Rachel'] == 1]
print(rachel_df)
```

```python
phoebe_df = df[df['Phoebe'] == 1]
print(phoebe_df)
```

### **Data Visualization**

```python
character_counts = {
    'Chandler': len(chandler_df),
    'Joey': len(joey_df),
    'Ross': len(ross_df),
    'Monica': len(monica_df),
    'Rachel': len(rachel_df),
    'Phoebe': len(phoebe_df)
}

labels = list(character_counts.keys())
sizes = list(character_counts.values())
```

```python
print(sizes)
```

```python
#pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Episode Titles Mentioning Each Character')
plt.show()
```

```python
#Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(character_counts.keys(), character_counts.values(), color='black')
plt.xlabel('Character')
plt.ylabel('Number of Titles')
plt.title('Number of Episode Titles Mentioning Each Character')
plt.show()
```

```python
#stacked bar chart
mentions = df[characters].sum()

plt.figure(figsize=(12, 8))
mentions.plot(kind='bar', stacked=True, color=['blue', 'orange', 'green', 'red', 'purple', 'brown'])
plt.xlabel('Character')
plt.ylabel('Number of Mentions')
plt.title('Overlapping Mentions in Episode Titles')
plt.show()
```

```python
#venn diagram
from matplotlib_venn import venn2, venn3

plt.figure(figsize=(8, 8))
venn2(subsets=(
    len(df[(df['Chandler'] == 1) & (df['Joey'] == 0)]),
    len(df[(df['Chandler'] == 0) & (df['Joey'] == 1)]),
    len(df[(df['Chandler'] == 1) & (df['Joey'] == 1)])),
    set_labels=('Chandler', 'Joey'))
plt.title('Venn Diagram: Chandler and Joey Mentions')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
venn2(subsets=(
    len(df[(df['Chandler'] == 1) & (df['Monica'] == 0)]),
    len(df[(df['Chandler'] == 0) & (df['Monica'] == 1)]),
    len(df[(df['Chandler'] == 1) & (df['Monica'] == 1)])),
    set_labels=('Chandler', 'Monica'))
plt.title('Venn Diagram: Chandler and Monica Mentions')
plt.show()
```

```python
#wordcloud
from wordcloud import WordCloud

chandler_titles = " ".join(df[df['Chandler'] == 1]['title'])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(chandler_titles)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Titles Mentioning Chandler')
plt.show()
```

```python
df['No_Character'] = df[characters].sum(axis=1) == 0
no_character_count = df['No_Character'].sum()
character_counts['No Character'] = no_character_count
```

```python
#pie chart
labels = list(character_counts.keys())
sizes = list(character_counts.values())

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Episode Titles Mentioning Each Character (Including None)')
plt.show()
```
